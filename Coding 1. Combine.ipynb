```python
#check input format before start
#title of input can't be changed unless needed

import pandas as pd

#define a function to read the csv files
def csv_read_deconv ():
    deconv_group1 = pd.read_csv ('0. input.csv')
    df1 = pd.DataFrame (deconv_group1)
    return df1
    
match = 1
match2 = 0 

#read the csv

deconv1 = csv_read_deconv()

#mass matching
#the number of samples can be added as needed, here shows the example for 4 samples
dimatch = pd.DataFrame(columns=['mass','RT','mass1','intensity1','RT_s1','RT_f1'])
dimatch['group'] = 0

#add start values: rank input file based on monoisotopic mass from high to low
dimatch.at[0,'mass1'] = deconv1.at[0,'Monoisotopic mass']
dimatch.at[0,'intensity1'] = deconv1.at[0,'Intensity']
dimatch.at[0,'RT_s1'] = deconv1.at[0,'RT_start']
dimatch.at[0,'RT_f1'] = deconv1.at[0,'RT_finish']
dimatch.at[0,'mass'] = deconv1.at[0,'Monoisotopic mass']
dimatch.at[0,'RT'] = deconv1.at[0,'RT_start']
dimatch.at[1,'mass1'] = deconv1.at[1,'Monoisotopic mass']
dimatch.at[1,'intensity1'] = deconv1.at[1,'Intensity']
dimatch.at[1,'RT_s1'] = deconv1.at[1,'RT_start']
dimatch.at[1,'RT_f1'] = deconv1.at[1,'RT_finish']
dimatch.at[1,'mass'] = deconv1.at[1,'Monoisotopic mass']
dimatch.at[1,'RT'] = deconv1.at[1,'RT_start']

for i in range(2,deconv1.shape[0]):
    sample = deconv1.at[i,'Sample number']
    mass = deconv1.at[i,'Monoisotopic mass']
    intensity = deconv1.at[i,'Intensity']
    RT_s = deconv1.at[i,'RT_start']
    RT_f = deconv1.at[i,'RT_finish']
    ave_mass = dimatch.at[match,'mass']
    ave_mass2 = dimatch.at[match2,'mass']
    ave_RT = dimatch.at[match,'RT']
    ave_RT2 = dimatch.at[match2,'RT']
    if (i % 100 == 0):
        print(i, ave_mass, ave_RT)
    ppm = abs((mass-ave_mass)/mass*1000000)
    ppm2 = abs((mass-ave_mass2)/mass*1000000)
    RT_d = abs(RT_s-ave_RT)
    RT_d2 = abs(RT_s-ave_RT2)
    if ppm <= 35 and RT_d <= 6.5:
        if sample == 1:
            dimatch.at[match,'mass1'] = mass
            dimatch.at[match,'intensity1'] = intensity
            dimatch.at[match,'RT_s1'] = RT_s
            dimatch.at[match,'RT_f1'] = RT_f
    elif ppm2 <=35 and RT_d2 <= 6.5:
        if sample == 1:
            dimatch.at[match2,'mass1'] = mass
            dimatch.at[match2,'intensity1'] = intensity 
            dimatch.at[match2,'RT_s1'] = RT_s
            dimatch.at[match2,'RT_f1'] = RT_f
    else:
        match2 = match
        match = match + 1
        if sample == 1:
            dimatch.at[match,'mass1'] = mass
            dimatch.at[match,'intensity1'] = intensity 
            dimatch.at[match,'RT_s1'] = RT_s
            dimatch.at[match,'RT_f1'] = RT_f
        dimatch.at[match,'mass'] = mass
        dimatch.at[match,'RT'] = RT_s
#dimatch.head(20)
dimatch.to_csv('1.Combined.csv')
print("Finished!")
```
