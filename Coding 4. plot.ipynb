```python
#Warning: change the TMT status range before run the coding!!!
import pandas as pd

#Read the csv files
def csv_read_deconv ():
    deconv_group = pd.read_csv ('3. TMT_grouped.csv')
    df = pd.DataFrame (deconv_group)
    return df
    
#Main loop heatmap 1
deconv = csv_read_deconv ()
heatmap1 = pd.DataFrame()
for i in range (0,deconv.shape[0]): 
    group = deconv.at[i,'group']
    sub_group = deconv.at[i,'sub_group']
    heatmap1.at[group,-100] = 0
    if deconv.at[i,'intensity1'] > 0:
        heatmap1.at[group,sub_group] = deconv.at[i,'intensity1']
heatmap1 = heatmap1.sort_index(axis = 1)
heatmap1.fillna(value=0,inplace=True)
for i in range (0,heatmap1.shape[0]):
    average = 0
    line = heatmap1.iloc[i]
    total = line.sum()
    heatmap1.at[i+1,'sum'] = total
    if total !=0:
        for j in range (-2,3):
            average = average + line[j] / total * j
    heatmap1.at[i+1,'ave'] = average
heatmap1.to_csv('5. Plot1.csv')
print('heatmap1 finished')

#Main loop heatmap 2
deconv = csv_read_deconv ()
heatmap2 = pd.DataFrame()
for i in range (0,deconv.shape[0]): 
    group = deconv.at[i,'group']
    sub_group = deconv.at[i,'sub_group']
    heatmap2.at[group,-100] = 0
    if deconv.at[i,'intensity2'] > 0:
        heatmap2.at[group,sub_group] = deconv.at[i,'intensity2']
heatmap2 = heatmap2.sort_index(axis = 1)
heatmap2.fillna(value=0,inplace=True)
for i in range (0,heatmap2.shape[0]):
    average = 0
    line = heatmap2.iloc[i]
    total = line.sum()
    heatmap2.at[i+1,'sum'] = total
    if total !=0:
        for j in range (-2,3):
            average = average + line[j] / total * j
    heatmap2.at[i+1,'ave'] = average
heatmap2.to_csv('5. Plot2.csv')
print('heatmap2 finished')

#Main loop heatmap 3
deconv = csv_read_deconv ()
heatmap3 = pd.DataFrame()
for i in range (0,deconv.shape[0]): 
    group = deconv.at[i,'group']
    sub_group = deconv.at[i,'sub_group']
    heatmap3.at[group,-100] = 0
    if deconv.at[i,'intensity3'] > 0:
        heatmap3.at[group,sub_group] = deconv.at[i,'intensity3']
heatmap3 = heatmap3.sort_index(axis = 1)
heatmap3.fillna(value=0,inplace=True)
for i in range (0,heatmap3.shape[0]):
    average = 0
    line = heatmap3.iloc[i]
    total = line.sum()
    heatmap3.at[i+1,'sum'] = total
    if total !=0:
        for j in range (-2,2):
            average = average + line[j] / total * j
    heatmap3.at[i+1,'ave'] = average
heatmap3.to_csv('5. Plot3.csv')
print('heatmap3 finished')

#Main loop heatmap 4
deconv = csv_read_deconv ()
heatmap4 = pd.DataFrame()
for i in range (0,deconv.shape[0]): 
    group = deconv.at[i,'group']
    sub_group = deconv.at[i,'sub_group']
    heatmap4.at[group,-100] = 0
    if deconv.at[i,'intensity4'] > 0:
        heatmap4.at[group,sub_group] = deconv.at[i,'intensity4']
heatmap4 = heatmap4.sort_index(axis = 1)
heatmap4.fillna(value=0,inplace=True)
for i in range (0,heatmap4.shape[0]):
    average = 0
    line = heatmap4.iloc[i]
    total = line.sum()
    heatmap4.at[i+1,'sum'] = total
    if total !=0:
        for j in range (-3,2):
            average = average + line[j] / total * j
    heatmap4.at[i+1,'ave'] = average
heatmap4.to_csv('5. Plot4.csv')
print('heatmap4 finished')
```
