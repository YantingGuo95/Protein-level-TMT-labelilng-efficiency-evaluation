```python
#Warning: change the TMT status range before run the coding!!!
import pandas as pd

#Read the csv files
def csv_read_deconv ():
    deconv_group = pd.read_csv ('4. TMT_subgrouped.csv')
    df = pd.DataFrame (deconv_group)
    return df
    
#Main loop heatmap 1
deconv = csv_read_deconv ()
heatmap1 = pd.DataFrame()
for i in range (0,deconv.shape[0]): 
    group = deconv.at[i,'group']
    sub_group = deconv.at[i,'sub_group']
    heatmap1.at[group,-100] = 0
    if deconv.at[i,'intensity1'] > 0:
        heatmap1.at[group,sub_group] = deconv.at[i,'intensity1']
heatmap1 = heatmap1.sort_index(axis = 1)
heatmap1.fillna(value=0,inplace=True)
for i in range (0,heatmap1.shape[0]):
    average = 0
    line = heatmap1.iloc[i]
    total = line.sum()
    heatmap1.at[i+1,'sum'] = total
    if total !=0:
        for j in range (-1,2):
            average = average + line[j] / total * j
    heatmap1.at[i+1,'ave'] = average
heatmap1.to_csv('5. Plot1.csv')
print('heatmap1 finished')
```
