```python
#Warning: change all blank to 0 in the input file before start!!!
#Main loop 2ï¼šFind isotopes with H and combine them

import pandas as pd
import numpy as np

#Read the csv files
def csv_read_deconv ():
    deconv_group1 = pd.read_csv ('1.Combined.csv')
    df1 = pd.DataFrame (deconv_group1)
    return df1

dimatch = csv_read_deconv()
match = 1
match2 = 0

for i in range(2,dimatch.shape[0]):
    mass_2 = dimatch.at[match2,'mass']
    RT_2 = dimatch.at[match2,'RT']
    mass_1 = dimatch.at[match,'mass']
    RT_1 = dimatch.at[match,'RT']
    mass = dimatch.at[i,'mass']
    RT = dimatch.at[i,'RT']
    ppm_plus = abs((mass + 1.007825 -mass_1)/mass*1000000)
    ppm_plus2 = abs((mass + 1.007825 -mass_2)/mass*1000000)
    RT_d = abs(RT_1-RT)
    RT_d2 = abs(RT_2-RT)
    if ppm_plus <= 35 and RT_d <= 6.5:
        if dimatch.at[i,'intensity1'] < dimatch.at[match,'intensity1']:
            dimatch.at[i,'mass1'] = dimatch.at[match,'mass1']
            dimatch.at[i,'intensity1'] = dimatch.at[match,'intensity1']
            dimatch.at[i,'RT_s1'] = dimatch.at[match,'RT_s1']
            dimatch.at[i,'RT_f1'] = dimatch.at[match,'RT_f1']            
        if dimatch.at[i,'intensity2'] < dimatch.at[match,'intensity2']:                                        
            dimatch.at[i,'mass2'] = dimatch.at[match,'mass2']
            dimatch.at[i,'intensity2'] = dimatch.at[match,'intensity2']
            dimatch.at[i,'RT_s2'] = dimatch.at[match,'RT_s2']
            dimatch.at[i,'RT_f2'] = dimatch.at[match,'RT_f2']
        if dimatch.at[i,'intensity3'] < dimatch.at[match,'intensity3']:                                                
            dimatch.at[i,'mass3'] = dimatch.at[match,'mass3']
            dimatch.at[i,'intensity3'] = dimatch.at[match,'intensity3']
            dimatch.at[i,'RT_s3'] = dimatch.at[match,'RT_s3']
            dimatch.at[i,'RT_f3'] = dimatch.at[match,'RT_f3']
        print(i,dimatch.at[match,'mass'])
        dimatch = dimatch.drop(labels=match)
        match = i
    elif ppm_plus2 <= 35 and RT_d2 <= 6.5:
        if dimatch.at[i,'intensity1'] < dimatch.at[match2,'intensity1']:
            dimatch.at[i,'mass1'] = dimatch.at[match2,'mass1']
            dimatch.at[i,'intensity1'] = dimatch.at[match2,'intensity1']
            dimatch.at[i,'RT_s1'] = dimatch.at[match2,'RT_s1']
            dimatch.at[i,'RT_f1'] = dimatch.at[match2,'RT_f1']            
        if dimatch.at[i,'intensity2'] < dimatch.at[match2,'intensity2']:                                             
            dimatch.at[i,'mass2'] = dimatch.at[match2,'mass2']
            dimatch.at[i,'intensity2'] = dimatch.at[match2,'intensity2']
            dimatch.at[i,'RT_s2'] = dimatch.at[match2,'RT_s2']
            dimatch.at[i,'RT_f2'] = dimatch.at[match2,'RT_f2']
        if dimatch.at[i,'intensity3'] < dimatch.at[match2,'intensity3']:                                               
            dimatch.at[i,'mass3'] = dimatch.at[match2,'mass3']
            dimatch.at[i,'intensity3'] = dimatch.at[match2,'intensity3']
            dimatch.at[i,'RT_s3'] = dimatch.at[match2,'RT_s3']
            dimatch.at[i,'RT_f3'] = dimatch.at[match2,'RT_f3']
        print(i,dimatch.at[match2,'mass'])
        dimatch = dimatch.drop(labels=match2)
        match2 = match
        match = i
    else:
        match2 = match
        match = i
dimatch.to_csv('2. Isotope filtered.csv')
```
