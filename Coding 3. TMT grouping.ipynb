```python
#change mas_diff if needed
#Examples:
#TMT0 mass: 224.1525
#TMT6plex mass: 229.1629

import pandas as pd

#Read the csv files
def csv_read_deconv ():
    deconv_group = pd.read_csv ('2. Isotope filtered.csv')
    df = pd.DataFrame (deconv_group)
    return df

#Main loop
deconv = csv_read_deconv ()
deconv['group'] = 0
TMT_group = 0
for i in range (0,deconv.shape[0]):
    if deconv.at[i,'group'] == 0:
        TMT_group = TMT_group + 1
        deconv.at[i,'group'] = TMT_group
    for j in range (i+1,deconv.shape[0]):
        mass_diff = abs(deconv.at[i,'mass'] - deconv.at[j,'mass'] - 224.1525) / deconv.at[i,'mass'] * 1000000
        RT_diff = abs(deconv.at[i,'RT'] - deconv.at[j,'RT'])
        #print(TMT_group,deconv.at[i,'mass'],deconv.at[j,'mass'],deconv.at[i,'RT'],deconv.at[j,'RT'],mass_diff,RT_diff)
        if (mass_diff < 200):
            if RT_diff < 6:
                deconv.at[j,'group'] = deconv.at[i,'group']
                continue
    if i%100 == 0:
        print(i)
deconv.head
deconv.to_csv('3. TMT_grouped2.csv')
print("Finished!")
```
